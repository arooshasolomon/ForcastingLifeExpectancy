```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df_life = pd.read_csv("/content/drive/MyDrive/life-expectancy.csv")
df_caloric = pd.read_csv("/content/drive/MyDrive/daily-per-capita-caloric-supply.csv")
df_gdp = pd.read_csv("/content/drive/MyDrive/gdp-per-capita-worldbank.csv")
df_health= pd.read_csv("/content/drive/MyDrive/total-healthcare-expenditure-gdp.csv")

df_life

dfs = {}
dfs["Life Expentancy"] = df_life
dfs["Caloric Supply"] = df_caloric
dfs["GDP"] = df_gdp
dfs["Healthcare Expenditure"] = df_health
```

```python
df_life_usa = df_life.loc[df_life['Code'] == 'USA']

plt.figure(figsize=(10,5))
plt.hist(df_life_usa['Period life expectancy at birth - Sex: total - Age: 0'],
         bins=10,
         color='skyblue',
         edgecolor='black')

plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.title('Histogram of Life Expectancy in the USA')
plt.grid(False)
plt.show()
```

```python
df_caloric_usa = df_caloric.loc[df_caloric['Code'] == 'USA']

plt.figure(figsize=(10,5))
plt.hist(df_caloric_usa['Daily calorie supply per person'],
         bins=10,
         color='skyblue',
         edgecolor='black')

plt.xlabel('Daily Caloric Supply Per Person')
plt.ylabel('Frequency')
plt.title('Distribution of Daily Caloric Supply in the USA')
plt.grid(False)
plt.show()
```

```python
df_gdp_usa = df_gdp.loc[df_gdp['Code'] == 'USA']

plt.figure(figsize=(10,5))
plt.hist(df_gdp_usa['GDP per capita, PPP (constant 2021 international $)'],
         bins=10,
         color='skyblue',
         edgecolor='black')

plt.xlabel('GDP per Capita (PPP, Constant 2021 $)')
plt.ylabel('Frequency')
plt.title('Distribution of GDP Per Capita in the USA')
plt.grid(False)
plt.show()
```

```python
df_health_usa = df_health.loc[df_health['Code'] == 'USA']

plt.figure(figsize=(10,5))
plt.hist(df_health_usa['Current health expenditure (CHE) as percentage of gross domestic product (GDP) (%)'],
         bins=10,
         color='skyblue',
         edgecolor='black')

plt.xlabel('Healthcare Expenditure as % of GDP')
plt.ylabel('Frequency')
plt.title('Distribution of Healthcare Expenditure in the USA')
plt.grid(False)
plt.show()
```
