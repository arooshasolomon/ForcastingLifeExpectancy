```python
#DF AGG

dfs = [dfs[key].loc[(dfs[key]['Year'] >= 2002) & (dfs[key]['Year'] <= 2021)] for key in dfs.keys()]

df1, df2, df3, df4 = dfs
```

```python
df1 = df1.loc[df1['Code'] == 'USA']
df2 = df2.loc[df2['Code'] == 'USA']
df3 = df3.loc[df3['Code'] == 'USA']
df4 = df4.loc[df4['Code'] == 'USA']
```

```python
dfs_new = [df1, df2, df3, df4]

merged_df = dfs[0]
for df in dfs[1:]:
    merged_df = merged_df.merge(df, on=["Entity", "Code", "Year"], how="outer")

# Reset index and display
merged_df.reset_index(drop=True, inplace=True)


merged_df
```

```python
import seaborn as sns
corr_matrix = merged_df.select_dtypes(include=['number']).corr()
corr_matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
```
